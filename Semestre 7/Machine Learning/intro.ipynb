**Question 7**: Try to build a classifier better than the random one, use the thresholding on one or more component. Choose the thresholds using the train dataset and compute the metrics on the test dataset.
    

```python
# En affichant le nuage de points précédent représentant l'ensemble des données et leurs classes respectives, on remarque deux groupes bien distincts de points : d'un côté les points rouges et de l'autre les points bleus. Afin de réaliser une séparation mathématique de ces deux ensembles de points, il est possible de tracer une droite (fonction affine) passant par exemple par les points (4.5, 2.25) et (6.5, 4). Cette fonction s'exprime sous la forme f(x) = ax + b où a est le coefficient directeur : (yA - yB)/(xA - xB) = (4.5 - 2.25)/(6.5 - 4.5) = 2.25/2 = 1.125, et b l'ordonnée à l'origine (fonction en x = 0) = f(0) => peut se calculer via f(4.5) et f(6.5). On sait que f(x) = ax + b = 1.143x + b. Or, f(4.5) = 2.25. Donc f(4.5) = 1.143 * 4.5 + b = 2.25. On conclut en disant que b = 2.25 - (1.143 * 4.5) = -2.893. Finalement, f(x) = 1.143x - 2.893

#
def threshold_curve_equation(x):
    return (1.125 * x) - 2.893

# prediction de la classe du point de coordonnées (x,y) : 0 si sur ou au-dessus de la fonction affine, 1 sinon.
def prediction_depending_on_threshold(x, y):
    if y >= threshold_curve_equation(x):
        return 0 # on retourne la classe 0 (point rouge) si le point qu'on veut prédire se trouve sur ou au-dessus de la droite séparatrice
    else:
        return 1 # on retourne la classe 1 (point bleu) si le point qu'on veut prédire se trouve en-dessous de la droite séparatrice

# la fonction ci-dessus peut se simplifier ainsi
def prediction_depending_on_threshold_simplified(x, y):
    return 0 if y >= threshold_curve_equation(x) else 1

X1Test = []
X2Test = []
for i in range(0, len(X_test)): # on parcourt toutes les données de test
    X1Test.append(X_test[i][0]) # on récupère les abscisses de nos points
    X2Test.append(X_test[i][1]) # on récupère les ordonnées de nos points

y_prediction_threshold = []
for i in range(0, len(X1Test)): # la condition d'arrêt aurait été identique avec X2Test
# on parcourt les paires (X1Test, X2Test) = [(X1Test[0], X2Test[0]), (X1Test[1], X2Test[1]), (X1Test[2], X2Test[2]), ..., (X1Test[N], X2Test[N])]
    y_prediction_threshold.append(prediction_depending_on_threshold_simplified(X1Test[i], X2Test[i]))

# affichage des métriques et de la matrice de confusion

from sklearn.metrics import accuracy_score, recall_score, f1_score, confusion_matrix

uniques_binary = np.unique(y_test)

print("accuracy of the binary classification", accuracy_score(y_test, y_prediction_threshold))

print("recall of the binary classification", recall_score(y_test, y_prediction_threshold, average=None))

print("f1_score of the binary classification", f1_score(y_test, y_prediction_threshold, average=None))

print("confusion matrix of the binary classification\n", confusion_matrix(y_test, y_prediction_threshold, labels=uniques_binary))
```

On se rend compte que représenter ici des points selon des coordonnées qui sont certaines valeurs des pixels n'a pas de sens. En effet, notre objectif premier est d'avoir une représentation graphique qui permettrait de voir une claire différence entre les positions des points appartenant à une image du chiffre 1, les positions des points appartenant à une image du chiffre 2, etc, afin de pouvoir aisément trouver une séparation entre les différents points du repère, permettant une prédiction efficace des chiffres présents sur des images reçues par l'algorithme. Ici, on voit que cela ne nous permet rien du tout, si ce n'est ne rien y comprendre.

On se demande alors comment faire pour avoir sur un repère des points qui permettraient d'identifier aisément chaque chiffre. Autrement dit, on veut trouver ce qui différencie clairement une image du chiffre 1 d'une image du chiffre 9. Il faut par ailleurs que ça soit quelque chose de "facilement" calculable.
Sachant que toutes les images de chiffres sont uniquement en nuance de gris, que le fond de chaque image est blanc et que les chiffres en eux-mêmes sont en noirs, on pourrait faire l'hypothèse que calculer le nombre de pixels "blancs"/"gris" (ou en tout cas, pas noirs) sur chaque image permettrait de différencier les différents chiffres (le fond d'un 9 aura peut-être moins de pixels blancs qu'un 1 par exemple). Mettons cela en application et voyons ce que cela nous donne en termes de représentation graphique.

```python
def count_light_pixel(x): # gray and whiter pixels
    return np.count_nonzero(x > 15) # compte le nombre de pixels dont la valeur (la nuance de gris) est supérieure à 15 (on considère qu'une valeur supérieure à 15 signifie que ce n'est pas un pixel noir)
```

```python
# appliquer la fonction  (merci Edgar pour la méthode que je ne connaissais pas)
X_train_white = np.apply_along_axis(count_light_pixel, 1, X_train)

# Supposons qu'on prend comme abscisse tous les X[i][1], c'est-à-dire la valeur d'indice 1 dans chaque image, et en ordonnée le nombre de pixels blancs/gris comptés dans l'image
plt.scatter(X2, X_train_white, c=['red' if y == 1 else 'blue' for y in y_train])
```

On ne peut toujours pas établir de séparation claire entre le chiffre 1 (pixels rouge) et le chiffre 9 (pixels bleu)...D'ailleurs, le fait d'utiliser X2 (première dimension de chaque image) comme coordonnée d'abscisse n'est pas une bonne idée ici car cela ne représente rien.

Cherchons quelque chose d'autre. Qu'est-ce qui permettrait de différencier le chiffre 1 du chiffre 9 dans un repère ? Une nouvelle idée qu'on peut avoir est de se dire que le chiffre 1 devrait avoir beaucoup de pixels noirs de manière verticale vers le milieu de l'image en abscisse par exemple, contrairement au chiffre 9 qui devrait en avoir moins. On rappelle que ce sont des images 8 * 8 dont les coordonnées sont enregistrées dans un seul axe, autrement dit, s'il y a une colonne verticale de pixels noirs représentant la barre du chiffre 1, elle sera sans doute vers le milieu de l'image en abscisse, c'est-à-dire vers le 4ème ou 5ème pixel sur 8 de chaque ligne de l'image.

On voit que sur l'image de l'extrait de dataset, le chiffre 1 a beaucoup de pixels noirs sur la 5ème colonne, contrairement au chiffre 9. On peut alors se dire qu'on peut se baser dans un premier temps sur le critère : "nombre de pixels noirs sur la 5ème colonne de l'image" pour différencier les chiffres 1 et 9.

```python
def count_dark_pixel(x): # gray and whiter pixels
    return np.count_nonzero(x < 3) # compte le nombre de pixels dont la valeur (la nuance de gris) est inférieure à 3 (on considère qu'une valeur inférieure à 3 signifie que ce n'est pas un pixel noir)
```
